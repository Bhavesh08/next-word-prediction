```python
f = re.compile('[][]')
```

```python
len(lines)
```

```python
s = [f.sub('',lines[x]) for x in range(len(lines))]
```

```python
words = [w.replace('\n','') for w in s]
```

```python
final_words = [y.replace('\'','')for y in words]
```

```python
final = [i.replace(' ' ' ','\n') for i in final_words]
```

```python
import functools
for k in range(len(final)):
    g = functools.reduce(lambda a,b: str(a)+str(" ")+str(b),final)
```

```python
g
```

```python
from keras.preprocessing.text import Tokenizer
token  = Tokenizer()
```

```python
token.fit_on_texts(final)
encoded = token.texts_to_sequences(final)
```

```python
output = [] 
  
# function used for removing nested  
# lists in python.  
def reemovNestings(l): 
    for i in l: 
        if type(i) == list: 
            reemovNestings(i) 
        else: 
            output.append(i) 
reemovNestings(encoded)
```

```python
print(output)
```

```python
len(output)
```

```python
print(token.word_index)
```

```python
voca_size = len(token.word_index)+1
voca_size
```

```python
sequences = list()
for i in range(1, len(output)):
	sequence = output[i-1:i+1]
	sequences.append(sequence)
print('Total Sequences: %d' % len(sequences))
```

```python
import numpy as np
from keras.utils import to_categorical
```

```python
sequences = np.array(sequences)

X, y = sequences[:,0],sequences[:,1]
# one hot encode outputs
y = to_categorical(y, num_classes=voca_size)
```

```python
print(y)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Embedding
```

```python
model = Sequential()
model.add(Embedding(voca_size, 20, input_length=1))
model.add(LSTM(311))
model.add(Dense(voca_size, activation='softmax'))
print(model.summary())
# compile network
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=1000, verbose=1)
```

```python
def generate_seq(model, tokenizer, seed_text, n_words):
	in_text, result = seed_text, seed_text
	# generate a fixed number of words
	for _ in range(n_words):
		# encode the text as integer
		encoded = token.texts_to_sequences([in_text])
		encoded = np.array(encoded)
		# predict a word in the vocabulary
		yhat = model.predict_classes(encoded, verbose=0)
		# map predicted word index to word
		out_word = ''
		for word, index in token.word_index.items():
			if index == yhat:
				out_word = word
				break
		# append to input
		in_text, result = out_word, result + ' ' + out_word
	return result
```

```python
print(generate_seq(model, token, 'gauri', 4))
```

```python
print(generate_seq(model, token, 'is', 5))
```
